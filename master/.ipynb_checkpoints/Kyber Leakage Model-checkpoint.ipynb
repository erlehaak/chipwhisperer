# Leakage model of kyber

```python
KYBER_N = 256
KYBER_Q = 3329
QINV = 3327
```

```python
def montgomery_reduction(a, q, qinv):
    m = (a * qinv) & ((1 << 32) - 1)
    u = (a + m * q) >> 32
    if u >= q:
        u -= q
    return u
```

```python
def doublebasemul_asm(poly0, poly1, poly2, poly3, zeta):
    q = 3329
    qinv = (1 << 32) // q

    tmp = (poly0 >> 16) * (poly1 >> 16)  # smultt
    tmp = montgomery_reduction(tmp, q, qinv)
    tmp2 = tmp * zeta
    tmp2 += (poly0 & 0xFFFF) * (poly1 & 0xFFFF)  # smlabb (smultb)
    tmp2 = montgomery_reduction(tmp2, q, qinv)

    tmp2_temp = ((poly0 & 0xFFFF) * (poly3 >> 16)) + ((poly0 >> 16) * (poly3 & 0xFFFF))  # smuadx
    tmp3 = montgomery_reduction(tmp2_temp, q, qinv)

    tmp = (tmp3 << 16) | (tmp & 0xFFFF)  # pkhtb

    zeta = -zeta

    tmp_temp = (poly2 >> 16) * (poly3 >> 16)  # smultt
    tmp_temp = montgomery_reduction(tmp_temp, q, qinv)
    tmp2 = tmp_temp * zeta
    tmp2 += (poly2 & 0xFFFF) * (poly3 & 0xFFFF)  # smlabb (smultb)
    tmp2 = montgomery_reduction(tmp2, q, qinv)

    tmp2_temp = ((poly2 & 0xFFFF) * (poly3 >> 16)) + ((poly2 >> 16) * (poly3 & 0xFFFF))  # smuadx
    tmp3 = montgomery_reduction(tmp2_temp, q, qinv)

    tmp2 = (tmp3 << 16) | (tmp2 & 0xFFFF)  # pkhtb

    return tmp, tmp2
```

```python
def reverse_bits(num):
    # num is an int16_t, so it should be between 0 and 0xFFFF
    num = (num & 0x5555) << 1 | (num & 0xAAAA) >> 1
    num = (num & 0x3333) << 2 | (num & 0xCCCC) >> 2
    num = (num & 0x0F0F) << 4 | (num & 0xF0F0) >> 4
    num = (num & 0x00FF) << 8 | (num & 0xFF00) >> 8
    return num

def reverse_hex(hex_int):
    # Convert int to hex string
    hex_string = hex(hex_int)[2:]  # remove '0x' prefix
    # Convert hex to binary
    binary_string = bin(int(hex_string, 16))[2:]
    # Pad with zeroes to make it multiple of 4 (each hex digit represents 4 binary digits)
    binary_string = binary_string.zfill(len(binary_string) + len(binary_string) % 4)
    # Reverse binary string
    reversed_binary_string = binary_string[::-1]
    # Convert reversed binary to int, then to hex
    reversed_hex_int = int(reversed_binary_string, 2)
    return reversed_hex_int
```
